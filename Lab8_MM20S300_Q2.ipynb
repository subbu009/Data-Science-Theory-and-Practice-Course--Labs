### Q2) Plot the surface as well as contours for the below two functions and visually verify the critical point(s).
 - (i) (w1 - 2 )^2 + (w2 - 2)^2 
 - (ii) w1^2 - w2^2 



 ### (i) The function is 
\begin{equation*}
J(w) = (w_1 - 2)^2 + (w_2 - 2)^2 
\end{equation*}

```python
def J1(w1,w2):
    return (w1-2)**2 + (w2-2)**2
```

```python
def J2(w1,w2):
    return w1**2 - w2**2
```

```python
x = np.arange(-10,14,0.1)
y = np.arange(-10,14,0.1)
w1,w2 = np.meshgrid(x,y)
#J1 = (w1-2)**2 + (w2-2)**2
fig = plt.figure(figsize=(10,4),dpi=100)
ax1 = fig.add_subplot(121, projection='3d',title = r'Surface plot : $J(w) = (w_1 - 2)^2 + (w_2 - 2)^2$ ',xlabel=r'$w_1$',ylabel=r'$w_2$')
ax1.plot_surface(w1,w2,J1(w1,w2))
ax2 = fig.add_subplot(122,title = r'Contour plot : $J(w) = (w_1 - 2)^2 + (w_2 - 2)^2$ ',xlabel=r'$w_1$',ylabel=r'$w_2$')
cs = plt.contour(w1,w2,J1(w1,w2))
plt.clabel(cs,inline=True,fontsize=10);
#fig.tight_layout()
```

### (ii) The function is 
\begin{equation*}
J(w) = w_1^2 - w_2^2 
\end{equation*}

```python
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)
w1,w2 = np.meshgrid(x,y)
#J2 = w1**2 - w2**2
fig = plt.figure(figsize=(10,4),dpi=100)
ax1 = fig.add_subplot(121, projection='3d',title = r'Surface plot : $J(w) = w_1^2 - w_2^2$ ',xlabel=r'$w_1$',ylabel=r'$w_2$')
ax1.plot_surface(w1,w2,J2(w1,w2))
ax2 = fig.add_subplot(122,title = r'Contour plot : $J(w) = w_1^2 - w_2^2$ ',xlabel=r'$w_1$',ylabel=r'$w_2$')
cs = plt.contour(w1,w2,J2(w1,w2))
plt.clabel(cs,inline=True,fontsize=10,levels=cs.levels);
```
