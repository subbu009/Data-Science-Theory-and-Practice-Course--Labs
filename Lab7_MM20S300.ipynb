### Q1) Generate a csv file as .txt consisting of the following columns - Students and marks. Populate them with at least five rows of data. Read and print them use python csv reader.

```python
import csv
```

```python
with open('marks.txt','w') as file:
    file.write('Students,Marks\n')
    file.write('Chandler,75\n')
    file.write('Joey,60\n')
    file.write('Ross,90\n')
    file.write('Monica,95\n')
    file.write('Pheobe,65\n')
    file.write('Rachel,70\n')
```

```python
with open('marks.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Header names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t\t{row[0]} scored {row[1]} marks in the test.')
            line_count += 1
```

### Q3) Plot a circle using its parametric form (r, theta). Also plot, x vs. theta and y vs. theta graphs. All graphs should appear in the same figure (use subplots()). Label all the plots appropriately. Give a suitable title for the figure. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r=1 #radius of the circle
theta = np.linspace(0,2*np.pi) 
x=np.cos(theta)
y=np.sin(theta)
```

```python
fig,ax = plt.subplots(3,1,figsize=(2,6),dpi=100)

ax[0].plot(x,y)
ax[0].set_aspect(1)
ax[0].set_title('Circle with radius 1 unit')
ax[0].set_xlabel("Radius (r)")
ax[0].set_ylabel("Theta")

ax[1].plot(theta,x)
ax[1].set_title('x is cosine of theta')
ax[1].set_xlabel("Theta")
ax[1].set_ylabel("X")

ax[2].plot(theta,y)
ax[2].set_title('y is sine of theta')
ax[2].set_xlabel("Theta")
ax[2].set_ylabel("Y")

fig.tight_layout()
```

### Q4) Plot the surface f(x,y) = (x - 10)^2 + (y - 10)^2. Also, generated the corresponding contour plot. Label the plots appropriately. Give a suitable title for the figure. 

```python
from matplotlib.figure import Figure
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.arange(0,21,0.01)
y=np.arange(0,21,0.01)
X,Y=np.meshgrid(x,y) #Forming MeshGrid
Z = (X-10)**2 + (Y-10) ** 2
fig = plt.figure(figsize=(5,5),dpi=100)
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F(X,Y)')
ax.set_title('Surface plot of F')
```

```python
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)
cp  = plt.contour(x, y, Z)
plt.clabel(cp, fontsize=8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F(X,y)')
ax.set_title('Contour plot of F')
plt.show()
```
